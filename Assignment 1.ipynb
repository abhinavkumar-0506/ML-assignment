```python
import numpy as np
arr=np.array([1,2,1,2,3,4,3,2,4,21,2,3,12,2,2,4])
most_freq = np.bincount(arr).argmax()
print("the most repeated value is ",most_freq)
indices = np.where(arr == most_freq)
print(" and their indices are ", indices)
```

```python
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print("sum of matrix" , gfg.sum())
print("row wise sum" , gfg.sum(axis=1))
print("coloumn wise sum" , gfg.sum(axis=0))
```

```python
import numpy as np
n_array=np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])
diag_sum = np.trace(n_array)
print("diagnol sum of matrix is " , diag_sum)
eigenvalue , eigenvector = np.linalg.eig(n_array)
print("eigen value of matrix is " , eigenvalue)
print("eigenvector of matrix is " , eigenvector)
inverse = np.linalg.inv(n_array)
print("inverse of matrix is " , inverse)
det = np.linalg.det(n_array)
print("determinant of matrix is " , det)
```

```python
import numpy as np

array = np.array([[1, -2, 3],
                  [-4, 5, -6]])

abs_array = np.abs(array)
print("i. Absolute values:\n", abs_array)

print("\nii. Percentiles:")

print("Flattened:", np.percentile(array, [25, 50, 75]))


print("Column-wise:\n", np.percentile(array, [25, 50, 75], axis=0))


print("Row-wise:\n", np.percentile(array, [25, 50, 75], axis=1))


print("\niii. Statistics:")


print("Flattened → Mean:", np.mean(array), 
      "Median:", np.median(array), 
      "Std Dev:", np.std(array))


print("Column-wise → Mean:", np.mean(array, axis=0))
print("Column-wise → Median:", np.median(array, axis=0))
print("Column-wise → Std Dev:", np.std(array, axis=0))

print("Row-wise → Mean:", np.mean(array, axis=1))
print("Row-wise → Median:", np.median(array, axis=1))
print("Row-wise → Std Dev:", np.std(array, axis=1))
```
